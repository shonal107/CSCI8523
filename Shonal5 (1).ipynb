```python
from tensorflow.keras.applications.inception_v3 import InceptionV3
from tensorflow.keras.applications.inception_v3 import preprocess_input
from tensorflow.keras.preprocessing import image
from tensorflow.keras.preprocessing.image import img_to_array
from sklearn.cluster import KMeans
import pandas as pd
import numpy as np
from tqdm import tqdm
import os
import shutil
import matplotlib.pyplot as plt 
```

```python
# Function to Extract features from the images
def image_feature(direc):
    model = InceptionV3(weights='imagenet', include_top=False)
    features = [];
    img_name = [];
    for i in tqdm(direc):
        fname='../data/resnet50v2_predicted_threshold70/river/'+i
        img=image.load_img(fname,target_size=(224,224))
        x = img_to_array(img)
        x=np.expand_dims(x,axis=0)
        x=preprocess_input(x)
        feat=model.predict(x)
        feat=feat.flatten()
        features.append(feat)
        img_name.append(i)
    return features,img_name
```

```python
img_path=os.listdir('../data/resnet50v2_predicted_threshold70/river')
img_features,img_name=image_feature(img_path)
```

```python
k = 2
clusters = KMeans(k, random_state = 40)
clusters.fit(img_features)
```

```python
image_cluster = pd.DataFrame(img_name,columns=['image'])
image_cluster["clusterid"] = clusters.labels_
image_cluster
```

```python
# Function to Extract features from the images
def image_feature(direc):
    model = InceptionV3(weights='imagenet', include_top=False)
    features = [];
    img_name = [];
    for i in tqdm(direc):
        fname='../data/shonal_test/river_hough/'+i
        img=image.load_img(fname,target_size=(224,224))
        x = img_to_array(img)
        x=np.expand_dims(x,axis=0)
        x=preprocess_input(x)
        feat=model.predict(x)
        feat=feat.flatten()
        features.append(feat)
        img_name.append(i)
    return features,img_name

img_path=os.listdir('../data/shonal_test/river_hough/')
img_features,img_name=image_feature(img_path)
```

```python
k = 2
clusters = KMeans(k, random_state = 40)
clusters.fit(img_features)

image_cluster = pd.DataFrame(img_name,columns=['image'])
image_cluster["clusterid"] = clusters.labels_
image_cluster
```

```python
cluster0 = []
cluster1 = []

for i in range(len(image_cluster)):
    if (image_cluster["clusterid"][i]==0):
        #print(str(i+1)+"/2726    cluster: 0    image: ", image_cluster["image"][i])
        cluster0.append(image_cluster["image"][i])
    else:
        #print(str(i+1)+"/2726    cluster: 1    image: ", image_cluster["image"][i])
        cluster1.append(image_cluster["image"][i])
```
